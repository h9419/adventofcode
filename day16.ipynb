```python
import numpy as np
from tqdm import trange
from enum import Enum
DEFAULT_FILENAME = "input16.txt"
def read_chars(filename = DEFAULT_FILENAME):
    res = []
    with open(filename, "r") as f:
        for li in f:
            res.append([*li.strip()])
    res = np.array(res)
    return res

class Dir(Enum):
  Right = 0
  Down = 1
  Left = 2
  Up = 3

values = read_chars()
```

```python
def walk(step, pos, dir):
    global values_visited
    global cursors
    if dir == Dir.Left:
        if pos[1] == 0:
            return
        values_visited[pos[0]][pos[1] - 1].append((step+1, Dir.Left))
        cursors.append((pos[0], pos[1] - 1, len(values_visited[pos[0]][pos[1] - 1]) - 1))
    elif dir == Dir.Right:
        if pos[1] == values.shape[1] - 1:
            return
        values_visited[pos[0]][pos[1] + 1].append((step+1, Dir.Right))
        cursors.append((pos[0], pos[1] + 1, len(values_visited[pos[0]][pos[1] + 1]) - 1))
    elif dir == Dir.Up:
        if pos[0] == 0:
            return
        values_visited[pos[0] - 1][pos[1]].append((step+1, Dir.Up))
        cursors.append((pos[0] - 1, pos[1], len(values_visited[pos[0] - 1][pos[1]]) - 1))
    elif dir == Dir.Down:
        if pos[0] == values.shape[0] - 1:
            return
        values_visited[pos[0] + 1][pos[1]].append((step+1, Dir.Down))
        cursors.append((pos[0] + 1, pos[1], len(values_visited[pos[0] + 1][pos[1]]) - 1))
def solve(start_cur, start_dir):
    global values_visited
    global cursors
    values_visited = [[[] for _ in range(values.shape[1])] for _ in range(values.shape[0])]
    values_visited[start_cur[0]][start_cur[1]].append((0, start_dir))
    cursors = [(start_cur[0],start_cur[1],0)]
    while len(cursors) > 0:
        cur = cursors.pop()
        hist = values_visited[cur[0]][cur[1]]
        _iter, _dir = hist[cur[2]]
        if len([elem for elem in hist[:cur[2]] if elem[1] == _dir]) > 0:
            continue
        current = values[cur[0]][cur[1]]
        if current == '.':
            walk(_iter, cur, _dir)
        elif current == '-':
            if _dir in [Dir.Right, Dir.Left]:
                walk(_iter, cur, _dir)
            else:
                walk(_iter, cur, Dir.Left)
                walk(_iter, cur, Dir.Right)
        elif current == '|':
            if _dir in [Dir.Up, Dir.Down]:
                walk(_iter, cur, _dir)
            else:
                walk(_iter, cur, Dir.Up)
                walk(_iter, cur, Dir.Down)
        elif current == '\\':
            _map = {
                Dir.Up: Dir.Left,
                Dir.Left: Dir.Up,
                Dir.Right: Dir.Down,
                Dir.Down: Dir.Right,
            }
            walk(_iter, cur, _map[_dir])
        elif current == '/':
            _map = {
                Dir.Up: Dir.Right,
                Dir.Right: Dir.Up,
                Dir.Left: Dir.Down,
                Dir.Down: Dir.Left,
            }
            walk(_iter, cur, _map[_dir])
    visited_count = 0
    for row in values_visited:
        for elem in row:
            visited_count += 1 if (len(elem) > 0) else 0
    return visited_count


print(f"Part 1: {solve((0, 0), Dir.Right)}")

sol2 = 0
for i in range(values.shape[1]):
    visited_count = solve((0, i), Dir.Down)
    if sol2 < visited_count:
        sol2 = visited_count
print(f"Part 2: {sol2}")
```
